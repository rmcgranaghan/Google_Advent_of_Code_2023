```python
txt_file = '/Users/ryanmc/Documents/Google_Advent_of_Code_2023/day2_input.txt'
f = open(txt_file,"r")
txt_data = f.read().split('\n')#.remove('')
f.close()

# int_data = [int(x) for x in txt_data]
# int_data
```

```python
red_limit = 12
green_limit = 13 
blue_limit = 14 
```

```python
def eval_trial(trial):
    red_limit = 12
    green_limit = 13 
    blue_limit = 14 
    
    trial = trial.split(',')
    for t in trial:
        
        text=""
        numbers=""
        res=[]
        for i in t:
            if(i.isdigit()):
                numbers+=i
            else:
                text+=i
        res.append(text)
        res.append(numbers)
    
         
        # # printing result
        # print("The tuple after the split of string and number : " + str(res))
    
        green_flag = 0
        blue_flag = 0
        red_flag = 0
        # check validity
        if 'green' in res[0]:
            if int(res[1]) > green_limit:
                green_flag = 1
        if 'blue' in res[0]:
            if int(res[1]) > blue_limit:
                blue_flag = 1
        if 'red' in res[0]:
            if int(res[1]) > red_limit:
                red_flag = 1
    
        if (green_flag == 1) | (blue_flag == 1) | (red_flag == 1):
            # print('game {} is impossible'.format(game))
            decision = -1
            return decision
        else:
            decision = 1

    
    return decision
```

```python

game_sums = 0
game_ids = []
for t in txt_data:
    print(t)
    txt_loop = re.split(r':|;',t)
    print(txt_loop)
    games_loop = re.sub('Game ','',txt_loop[0])
    print(games_loop)
    txt_loop = txt_loop[1:]

    game_decision = 1
    for trial_loop in txt_loop:
        print('----> {}'.format(trial_loop))
        trial_decision = eval_trial(trial_loop)
        if trial_decision == -1:
            game_decision = -1
            print('game {} is impossible...continuing\n\n\n'.format(games_loop))
            break
    if game_decision == 1:
        game_sums += int(games_loop)
        game_ids.append(int(games_loop))
            
print('answer to day 2 part 1 = {}'.format(game_sums))
```

```python
game_sums
```

```python
def eval_trial_mins(trial):
    
    trial = trial.split(',')

    for t in trial:
        
        text=""
        numbers=""
        res=[]
        for i in t:
            if(i.isdigit()):
                numbers+=i
            else:
                text+=i
        res.append(text)
        res.append(numbers)
    
         
        # # printing result
        # print("The tuple after the split of string and number : " + str(res))
    
        # check validity
        if 'green' in res[0]:
            greens = int(res[1])

        if 'blue' in res[0]:
            blues = int(res[1])
        if 'red' in res[0]:
            reds = int(res[1])

    try:
        greens
    except:
        greens = np.nan

    try:
        blues
    except:
        blues = np.nan

    try:
        reds
    except:
        reds = np.nan

    
    return greens,blues,reds
```

```python

game_mins_power = []
for t in txt_data:
    print(t)
    txt_loop = re.split(r':|;',t)
    print(txt_loop)
    games_loop = re.sub('Game ','',txt_loop[0])
    print(games_loop)
    txt_loop = txt_loop[1:]

    game_decision = 1
    game_greens = []
    game_blues = []
    game_reds = []
    
    for trial_loop in txt_loop:
        print('----> {}'.format(trial_loop))
        trial_greens,trial_blues,trial_reds = eval_trial_mins(trial_loop)
        
        game_greens.append(trial_greens)
        game_blues.append(trial_blues)
        game_reds.append(trial_reds)
    print('this games green = {}, blue = {}, red = {}'.format(game_greens,game_blues,game_reds))
        
    game_mins_power.append(np.nanmax(game_greens) * np.nanmax(game_blues) * np.nanmax(game_reds))
    print('\n\n\n')

            
print('answer to day 2 part 2 = {}'.format( np.sum(game_mins_power) ))
```

```python
np.sum(game_mins_power)
```

```python
special_char="'[@_!$%^&*()<>-`?;,<>+=/\|}{~:]#'"
```

```python
def find_num_first_line(loop_num_inds,txt_data,r):
    if (loop_num_inds[0] > 0) & (loop_num_inds[1] < len(txt_data[r])-1):
        # Check all surrounding symbols
        for x in range(loop_num_inds[0]-1,loop_num_inds[1]+1):
            for y in range(r,r+2):
                if txt_data[y][x] in special_char:
                    stored_numbers.append(int(txt_data[r][loop_num_inds[0]:loop_num_inds[1]]))
                    print('found a number {} to save among {}'.format(txt_data[r][loop_num_inds[0]:loop_num_inds[1]],txt_data[r]))
                    return int(txt_data[r][loop_num_inds[0]:loop_num_inds[1]])
    elif (loop_num_inds[0] == 0):
        # Check all surrounding symbols
        for x in range(loop_num_inds[0],loop_num_inds[1]+1):
            for y in range(r,r+2):
                if txt_data[y][x] in special_char:
                    stored_numbers.append(int(txt_data[r][loop_num_inds[0]:loop_num_inds[1]]))
                    print('found a number {} to save among {}'.format(txt_data[r][loop_num_inds[0]:loop_num_inds[1]],txt_data[r]))
                    return int(txt_data[r][loop_num_inds[0]:loop_num_inds[1]])
    elif (loop_num_inds[1] == len(txt_data[r])-1):
        # Check all surrounding symbols
        for x in range(loop_num_inds[0]-1,loop_num_inds[1]):
            for y in range(r,r+2):
                if txt_data[y][x] in special_char:
                    stored_numbers.append(int(txt_data[r][loop_num_inds[0]:loop_num_inds[1]]))
                    print('found a number {} to save among {}'.format(txt_data[r][loop_num_inds[0]:loop_num_inds[1]],txt_data[r]))
                    return int(txt_data[r][loop_num_inds[0]:loop_num_inds[1]])
    return np.nan

def find_num_last_line(loop_num_inds,txt_data,r):

     # last line case

    if (loop_num_inds[0] > 0) & (loop_num_inds[1] < len(txt_data[r])-1):
        # Check all surrounding symbols
        for x in range(loop_num_inds[0]-1,loop_num_inds[1]+1):
            for y in range(r-1,r+1):
                if txt_data[y][x] in special_char:
                    stored_numbers.append(int(txt_data[r][loop_num_inds[0]:loop_num_inds[1]]))
                    print('found a number {} to save among {}'.format(txt_data[r][loop_num_inds[0]:loop_num_inds[1]],txt_data[r]))
                    return int(txt_data[r][loop_num_inds[0]:loop_num_inds[1]])
    elif (loop_num_inds[0] == 0):
        # Check all surrounding symbols
        for x in range(loop_num_inds[0],loop_num_inds[1]+1):
            for y in range(r-1,r+1):
                if txt_data[y][x] in special_char:
                    stored_numbers.append(int(txt_data[r][loop_num_inds[0]:loop_num_inds[1]]))
                    print('found a number {} to save among {}'.format(txt_data[r][loop_num_inds[0]:loop_num_inds[1]],txt_data[r]))
                    return int(txt_data[r][loop_num_inds[0]:loop_num_inds[1]])
    elif (loop_num_inds[1] == len(txt_data[r])-1):
        # Check all surrounding symbols
        for x in range(loop_num_inds[0]-1,loop_num_inds[1]):
            for y in range(r-1,r+1):
                if txt_data[y][x] in special_char:
                    stored_numbers.append(int(txt_data[r][loop_num_inds[0]:loop_num_inds[1]]))
                    print('found a number {} to save among {}'.format(txt_data[r][loop_num_inds[0]:loop_num_inds[1]],txt_data[r]))
                    return int(txt_data[r][loop_num_inds[0]:loop_num_inds[1]])
    return np.nan


def find_num_middle_line(loop_num_inds,txt_data,r):

    # Most regular case
    # elif (loop_num_inds[1] > 0) & (loop_num_inds[1] < len(txt_data[r])):
    if (loop_num_inds[0] > 0) & (loop_num_inds[1] < len(txt_data[r])-1):
        # Check all surrounding symbols
        for x in range(loop_num_inds[0]-1,loop_num_inds[1]+1):
            for y in range(r-1,r+2):
                # print('x = {}, y = {}'.format(x,y))
                if txt_data[y][x] in special_char:
                    stored_numbers.append(int(txt_data[r][loop_num_inds[0]:loop_num_inds[1]]))
                    print('found a number {} to save among {}'.format(txt_data[r][loop_num_inds[0]:loop_num_inds[1]],txt_data[r]))
                    return int(txt_data[r][loop_num_inds[0]:loop_num_inds[1]])
    elif (loop_num_inds[0] == 0):
        # Check all surrounding symbols
        for x in range(loop_num_inds[0],loop_num_inds[1]+1):
            for y in range(r-1,r+2):
                if txt_data[y][x] in special_char:
                    stored_numbers.append(int(txt_data[r][loop_num_inds[0]:loop_num_inds[1]]))
                    print('found a number {} to save among {}'.format(txt_data[r][loop_num_inds[0]:loop_num_inds[1]],txt_data[r]))
                    return int(txt_data[r][loop_num_inds[0]:loop_num_inds[1]])
    elif (loop_num_inds[1] == len(txt_data[r])-1):
        # Check all surrounding symbols
        for x in range(loop_num_inds[0]-1,loop_num_inds[1]):
            for y in range(r-1,r+2):
                if txt_data[y][x] in special_char:
                    stored_numbers.append(int(txt_data[r][loop_num_inds[0]:loop_num_inds[1]]))
                    print('found a number {} to save among {}'.format(txt_data[r][loop_num_inds[0]:loop_num_inds[1]],txt_data[r]))
                    return int(txt_data[r][loop_num_inds[0]:loop_num_inds[1]])

    return np.nan
```

```python

stored_numbers = []
sum_loop = 0

for r in range(len(txt_data)):
    print(txt_data[r])

    for m in re.finditer("\d+", txt_data[r]):
        loop_num_inds = m.span()
        print('r = {}, loop indices = {}'.format(r, loop_num_inds))           

            
        # first line case
        if (r == 0):
            num_temp = find_num_first_line(loop_num_inds,txt_data,r)
  

        
        # last line case
        elif (r == len(txt_data)-1):
            num_temp = find_num_last_line(loop_num_inds,txt_data,r)

        # Most regular case
        else:
            num_temp = find_num_middle_line(loop_num_inds,txt_data,r)

        stored_numbers.append(num_temp)
        if not math.isnan(num_temp):
            sum_loop += num_temp
```

```python
import math as m, re

board = list(open('day3_input.txt'))
chars = {(r, c): [] for r in range(140) for c in range(140)
# chars = {(r, c): [] for r in range(10) for c in range(10)
                    if board[r][c] not in '01234566789.'}

for r, row in enumerate(board):
    for n in re.finditer(r'\d+', row):
        edge = {(r, c) for r in (r-1, r, r+1)
                       for c in range(n.start()-1, n.end()+1)}

        for o in edge & chars.keys():
            chars[o].append(int(n.group()))

print(sum(sum(p)    for p in chars.values()),
      sum(m.prod(p) for p in chars.values() if len(p)==2))
```
