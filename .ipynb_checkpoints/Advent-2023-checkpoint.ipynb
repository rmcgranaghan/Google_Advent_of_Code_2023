# [Day 2](https://adventofcode.com/2023/day/2)

```python
txt_file = '/Users/ryanmc/Documents/Google_Advent_of_Code_2023/day2_input.txt'
f = open(txt_file,"r")
txt_data = f.read().split('\n')#.remove('')
f.close()

# int_data = [int(x) for x in txt_data]
# int_data
```

```python
txt_data
```

```python
red_limit = 12
green_limit = 13 
blue_limit = 14 
```

```python
def eval_trial(trial):
    red_limit = 12
    green_limit = 13 
    blue_limit = 14 
    
    trial = trial.split(',')
    for t in trial:
        
        text=""
        numbers=""
        res=[]
        for i in t:
            if(i.isdigit()):
                numbers+=i
            else:
                text+=i
        res.append(text)
        res.append(numbers)
    
         
        # # printing result
        # print("The tuple after the split of string and number : " + str(res))
    
        green_flag = 0
        blue_flag = 0
        red_flag = 0
        # check validity
        if 'green' in res[0]:
            if int(res[1]) > green_limit:
                green_flag = 1
        if 'blue' in res[0]:
            if int(res[1]) > blue_limit:
                blue_flag = 1
        if 'red' in res[0]:
            if int(res[1]) > red_limit:
                red_flag = 1
    
        if (green_flag == 1) | (blue_flag == 1) | (red_flag == 1):
            # print('game {} is impossible'.format(game))
            decision = -1
            return decision
        else:
            decision = 1

    
    return decision
```

```python

game_sums = 0
game_ids = []
for t in txt_data:
    print(t)
    txt_loop = re.split(r':|;',t)
    print(txt_loop)
    games_loop = re.sub('Game ','',txt_loop[0])
    print(games_loop)
    txt_loop = txt_loop[1:]

    game_decision = 1
    for trial_loop in txt_loop:
        print('----> {}'.format(trial_loop))
        trial_decision = eval_trial(trial_loop)
        if trial_decision == -1:
            game_decision = -1
            print('game {} is impossible...continuing\n\n\n'.format(games_loop))
            break
    if game_decision == 1:
        game_sums += int(games_loop)
        game_ids.append(int(games_loop))
            
print('answer to day 2 part 1 = {}'.format(game_sums))
```

```python
def eval_trial_mins(trial):
    
    trial = trial.split(',')

    for t in trial:
        
        text=""
        numbers=""
        res=[]
        for i in t:
            if(i.isdigit()):
                numbers+=i
            else:
                text+=i
        res.append(text)
        res.append(numbers)
    
         
        # # printing result
        # print("The tuple after the split of string and number : " + str(res))
    
        # check validity
        if 'green' in res[0]:
            greens = int(res[1])

        if 'blue' in res[0]:
            blues = int(res[1])
        if 'red' in res[0]:
            reds = int(res[1])

    try:
        greens
    except:
        greens = np.nan

    try:
        blues
    except:
        blues = np.nan

    try:
        reds
    except:
        reds = np.nan

    
    return greens,blues,reds
```

```python

game_mins_power = []
for t in txt_data:
    print(t)
    txt_loop = re.split(r':|;',t)
    print(txt_loop)
    games_loop = re.sub('Game ','',txt_loop[0])
    print(games_loop)
    txt_loop = txt_loop[1:]

    game_decision = 1
    game_greens = []
    game_blues = []
    game_reds = []
    
    for trial_loop in txt_loop:
        print('----> {}'.format(trial_loop))
        trial_greens,trial_blues,trial_reds = eval_trial_mins(trial_loop)
        
        game_greens.append(trial_greens)
        game_blues.append(trial_blues)
        game_reds.append(trial_reds)
    print('this games green = {}, blue = {}, red = {}'.format(game_greens,game_blues,game_reds))
        
    game_mins_power.append(np.nanmax(game_greens) * np.nanmax(game_blues) * np.nanmax(game_reds))
    print('\n\n\n')

            
print('answer to day 2 part 2 = {}'.format( np.sum(game_mins_power) ))
```
